```python
mat = np.zeros([100,100,3]) # [n,m,k]
for n in range(100):
    for m in range(100): 
        mat[n,m,:] = np.random.rand()

plt.imshow(mat, cmap = "gray", label = "test")
plt.colorbar()
plt.show()
```

```python
n, m, k = 100,100,3
mat = np.zeros([n,m,k]) # [n,m,k]
for n_ in range(n):
    for m_ in range(m): 
        mat[n_,m_,:] = np.random.rand(k)*0.9

plt.imshow(mat)
plt.colorbar()
plt.show()
```

```python
# bestimmte Anpassungen fr Matrizen
import numpy as np
mat = np.zeros([10, 10])
for i in range(10):
    for j in range(10):
        if i == j: # Hauptdiagonale
            mat[i, j] = 1
print(mat)
```
