```python
np.random.seed(0)

N = 50
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)
colors2 = ["red","blue","green","yellow"]
ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors2, alpha=0.5)

plt.show()
```

```python
h = 6.62607015e-34 # Planck constant
T = 300 # temperature in Kelvin
c = 299792458 # speed of light
kB = 1.380649e-23 # Boltzmann constant

spectralDensity = lambda nu : ((2 * h * nu**3) / (c**2)) / \
(math.exp((h * nu) / (kB * T)) - 1)

X = [x for x in range(1, int(1e+14), int(1e+10))]
Y = [spectralDensity(x) for x in X]

plt.title("Schwarzk√∂rperstrahlung")
plt.xlabel("Strahlungsfrequenz in Hz")
plt.ylabel("Intensit√§t in W/m2")

plt.plot(X, Y)
plt.show()
```
