## Aufgabe 1:
Schreibe ein Programm, das ein Bild generiert, welcher nur aus zufälligem Rauschen besteht.
Erwartete Bildgröße:
- Eingabe: Breite = 256, Höhe = 256
- Ausgabe: Ein 256x256 Bild mit zufälligem Rauschen.

```python
import numpy as np
import matplotlib.pyplot as plt

n, m, k = 256, 256, 3
mat = np.zeros([n,m,k])
for n_ in range(n):
    for m_ in range(m):
        mat[n_,m_,:] = np.random.rand(k)

plt.imshow(mat)
```

## Aufgabe 2:
Erstelle ein Programm, das eine $n \times n$ Matrix mit $n$ Zeilen und $m$ Spalten erstellt.
Dabei soll sein:
- $n \land m$ = 5
- es sollen Einsen auf folgenden Positionen sein: [(0, 0), (1,2), (3,4)]

```python
import numpy as np

mat = np.zeros([5,5])
for i in range(5):
    for j in range(5):
        if i == 0  and j == 0:
            mat[i,j] = 1
        if i == 1 and j == 2:
            mat[i,j] = 1
        if i == 3 and j == 4:
            mat[i,j] = 1
print(mat)
```

```python
# alternativ
import numpy as np
n,m = 5,5
mat = np.zeros([n,m])

positions = [(0,0),(1,2),(3,4)]

for pos in positions:
    mat[pos] = 1
print(mat)
```

## Aufgabe 3:
Definiere eine Klasse Rechteck, die folgende Funktionalität bietet:
- Attribute: breite, hoehe
- Methoden: fläche(), skalieren(faktor)

Als Test soll ein Rechteck die breite = 4 und höhe = 3 haben. Diese Fläche soll einmal mit 2 skaliert werden

```python
class Rechteck:
    def __init__(self, breite, hoehe):
        self.breite = breite
        self.hoehe = hoehe

    def fläche(self):
        return self.breite * self.hoehe

    def skalieren(self, faktor):
        self.breite *= faktor
        self.hoehe *= faktor

mein_rechteck = Rechteck(4, 3) 

print(f"Fläche: {mein_rechteck.fläche()}") 

mein_rechteck.skalieren(2)
print(f"Neue Fläche: {mein_rechteck.fläche()}")
```
