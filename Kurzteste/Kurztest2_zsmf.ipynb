```python
# Beispiel einer Klasse
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        real_sum = self.real + other.real
        imag_sum = self.imag + other.imag
        return ComplexNumber(real_sum, imag_sum)

c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(4, 5)
result = c1 + c2
print(f"Result: {result.real} + {result.imag}i")
```

```python
# bestimmte Anpassungen f√ºr Matrizen
import numpy as np
mat = np.zeros([10, 10])
for i in range(10):
    for j in range(10):
        if i == j: # Hauptdiagonale
            mat[i, j] = 1
print(mat)
```

Diese Matrix entpricht im mathematische Sinne dem Kronecker-Delta:
$$\delta_{ij} = \begin{cases} 1, \text{ falls } i=j \\ 0, \text{ falls } i  \ne j \end{cases}

```python
import numpy as np
mat = np.zeros([10, 10])
for i in range(10):
    for j in range(10):
        if i+1 == j: # Nebendiagonale oben
            mat[i, j] = 1
        if i-1 == j: # Nebendiagonale unten
            mat[i, j] = 1
print(mat)
```

```python
# 2D-plots
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
data1 = np.random.normal(loc=50, scale=5, size=1000)  # Example: Particle energies
x = np.linspace(0, 10, 100)  # Example: Time
y1 = np.sin(x)               # Example: Sine wave for oscillations
y2 = np.cos(x)               # Example: Cosine wave for oscillations
data2 = np.random.rand(500, 2) * 10  # Example: Positions of particles
categories = ['Protons', 'Neutrons', 'Electrons']
counts = [40, 35, 25]  # Example: Particle counts
data3 = [np.random.rand(50) for _ in range(3)]  # Example: Results from experiments

# Create subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Histogram
axs[0, 0].hist(data1, bins=20, color='blue', alpha=0.7)
axs[0, 0].set_title('Histogram: Particle Energies')
axs[0, 0].set_xlabel('Energy (MeV)')
axs[0, 0].set_ylabel('Count')

# 2D Histogram
axs[0, 1].hist2d(data2[:, 0], data2[:, 1], bins=20, cmap='viridis')
axs[0, 1].set_title('2D Histogram: Particle Positions')
axs[0, 1].set_xlabel('x-position')
axs[0, 1].set_ylabel('y-position')

# Pie Chart
axs[1, 0].pie(counts, labels=categories, autopct='%1.1f%%', colors=['red', 'green', 'blue'])
axs[1, 0].set_title('Pie Chart: Particle Composition')

# Box Plot
axs[1, 1].boxplot(data3, vert=True, patch_artist=True)
axs[1, 1].set_title('Box Plot: Experiment Results')
axs[1, 1].set_xticklabels(['Exp 1', 'Exp 2', 'Exp 3'])
axs[1, 1].set_ylabel('Value')

# Scatter Plot
axs[2, 0].scatter(data2[:, 0], data2[:, 1], c=data2[:, 1], cmap='plasma', alpha=0.7)
axs[2, 0].set_title('Scatter Plot: Particle Positions')
axs[2, 0].set_xlabel('x')
axs[2, 0].set_ylabel('y')

# Stack Plot
axs[2, 1].stackplot(x, y1, y2, labels=['sin(x)', 'cos(x)'], alpha=0.8)
axs[2, 1].set_title('Stack Plot: Oscillations')
axs[2, 1].set_xlabel('Time (s)')
axs[2, 1].legend()

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Data for 3D plots
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Example: Wave function

x_bar = np.arange(1, 11)
y_bar = np.random.randint(1, 10, size=10)
z_bar = np.zeros(10)
dx = dy = np.ones(10)
dz = np.random.randint(1, 10, size=10)  # Example: 3D histogram bin heights

fig = plt.figure(figsize=(16, 12))

# 3D Line Plot
ax1 = fig.add_subplot(231, projection='3d')
ax1.plot(x, np.sin(x), np.cos(x))
ax1.set_title('3D Line Plot: Oscillations')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.set_zlabel('cos(x)')

# 3D Scatter Plot
ax2 = fig.add_subplot(232, projection='3d')
ax2.scatter(X.flatten(), Y.flatten(), Z.flatten(), c=Z.flatten(), cmap='viridis')
ax2.set_title('3D Scatter: Wave Function')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

# 3D Wireframe Plot
ax3 = fig.add_subplot(233, projection='3d')
ax3.plot_wireframe(X, Y, Z, color='black')
ax3.set_title('3D Wireframe: Wave Function')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z')

# 3D Surface Plot
ax4 = fig.add_subplot(234, projection='3d')
surface = ax4.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none')
ax4.set_title('3D Surface: Wave Function')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_zlabel('z')
fig.colorbar(surface, ax=ax4, shrink=0.5, aspect=10)

# 3D Bar Plot
ax5 = fig.add_subplot(235, projection='3d')
ax5.bar3d(x_bar, y_bar, z_bar, dx, dy, dz, shade=True)
ax5.set_title('3D Bar Plot: Histogram Bins')
ax5.set_xlabel('x')
ax5.set_ylabel('y')
ax5.set_zlabel('z')

plt.tight_layout()
plt.show()
```
