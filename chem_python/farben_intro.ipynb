```python
import matplotlib.pyplot as plt
import numpy as np
import colorsys
from tqdm import tqdm
```

```python
alle_farben = np.zeros([256**3,3])
i = 0
for r in tqdm(range(256)):
    for g in range(256):
        for b in range(256):
            alle_farben[i,:] = [r/255,g/255,b/255]
            i += 1

square = np.reshape(alle_farben, (4096, 4096, 3))

plt.imshow(square)
```
