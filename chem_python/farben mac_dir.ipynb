```python
from PIL import Image
import numpy as np
from copy import deepcopy
import matplotlib.pyplot as plt
import colorsys
from tqdm import tqdm

p = "/Users/mikariesterer/Downloads/IMG_1849.png"
image = Image.open(p)
image_array = np.array(image)
```

```python
img = np.array(deepcopy(image_array[1500:3500,800:3200,:])/256)
```

```python
plt.imshow(img, origin='upper')
```

```python
hsv = np.zeros_like(img)
for i in tqdm(range(img.shape[0])):
    for j in range(img.shape[1]):
        hsv[i,j,:] = deepcopy(colorsys.rgb_to_hsv(deepcopy(img[i,j,0]),deepcopy(img[i,j,1]),deepcopy(img[i,j,2])))
```
